```python
label_pth='DeepSign/labels'
if not os.path.isdir(label_pth):
    os.makedirs(label_pth)
```

```python
for index, row in df.iterrows():
    print(row[0])
    line = list(row[1:5])
    clas = str(row[5])
    res = signs.index(clas)
    line.append(res)
    line = intlist2txt(line)
    print(line)
    #line = list(df.iloc[0][1:5])
  
    pre, ext = os.path.splitext(row[0])
    txt_name = os.path.join(label_pth, pre + ".txt")
    #print(txt_name)
    file1 = open(txt_name, "a")  # append mode
    file1.write(line + "\n")
    file1.close()
```
